```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from googleapiclient.discovery import build
```

```python
def get_channel_videos(channel_id):
    video_ids = []
    next_page_token = None

    while True:
        res = youtube.search().list(part='snippet', channelId=channel_id,
                                    maxResults=50, pageToken=next_page_token,
                                    order='date').execute()
        
        for item in res['items']:
            if item['id']['kind'] == 'youtube#video':
                video_ids.append(item['id']['videoId'])
        
        next_page_token = res.get('nextPageToken')
        if not next_page_token:
            break

    return video_ids

video_ids = get_channel_videos(channel_id)
print("Total videos found:", len(video_ids))
```

```python
def get_video_stats(video_ids):
    video_data = []

    for i in range(0, len(video_ids), 50):  # YouTube API allows max 50 IDs per request
        res = youtube.videos().list(
            part='snippet,statistics',
            id=','.join(video_ids[i:i+50])
        ).execute()

        for video in res['items']:
            snippet = video['snippet']
            stats = video['statistics']
            video_data.append({
                'title': snippet['title'],
                'published_at': snippet['publishedAt'],
                'views': int(stats.get('viewCount', 0)),
                'likes': int(stats.get('likeCount', 0)),
                'comments': int(stats.get('commentCount', 0))
            })
    
    return video_data

video_stats = get_video_stats(video_ids)
df = pd.DataFrame(video_stats)
df['published_at'] = pd.to_datetime(df['published_at'])
df.head()
```

```python
# Remove timezone info to avoid Excel error
df['published_at'] = df['published_at'].dt.tz_localize(None)

# Now save to Excel safely
df.to_excel('YouTube_Engagement.xlsx', index=False)
```

```python
df['month'] = df['published_at'].dt.to_period('M')
monthly = df.groupby('month')[['views', 'likes', 'comments']].sum().reset_index()

monthly.plot(x='month', y=['views', 'likes', 'comments'], kind='line', figsize=(12,6), marker='o')
plt.title('Monthly Engagement Trend')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['weekday'] = df['published_at'].dt.day_name()
weekday_avg = df.groupby('weekday')[['views', 'likes', 'comments']].mean().sort_values('views', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(data=weekday_avg.reset_index(), x='weekday', y='views', palette='viridis')
plt.title('Average Views by Weekday')
plt.ylabel('Average Views')
plt.xticks(rotation=45)
plt.show()
```
